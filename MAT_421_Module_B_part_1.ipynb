Question: Find the linear interpolation at x1 = 0.5 and x2 = 1.5 for the data x = [0,1,2], y = [5,2,4].

Import the necessary libararies.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

Create a function that follows the linear interpolation formula.

```python
x,y = [0,1,2],[5,2,4]

def inter(yi,yii,xi,xii,x):
  result = yi+((yii-yi)*(x-xi))/(xii-xi)
  return(result)

s = ("The predicted interpolated value for x1 is " + repr(inter(5,2,0,1,0.5)) + " and the predicted interpolated value for x2 is " + repr(inter(2,4,1,2,1.5)) + "." )
print(s)
```

We can verify this with a function from scipy.

```python
function = interp1d(x,y)
y_hat1 = function(0.5)
y_hat2 = function(1.5)
s = ("The actual interpolated value for x1 is " + repr(y_hat1) + " and the actual interpolated value for x2 is " + repr(y_hat2) + "." )
print(s)
```

By using the formula for linear interpolation, the predicted y values for x1 = 0.5 and x2 = 1.5 were obtained.

Question: Plot the previous problem and the interpolated values.

```python
plt.figure(figsize = (10,10))
plt.plot(x, y, '-ob')
plt.plot(0.5, y_hat1, 'ro')
plt.plot(1.5, y_hat2, 'ro')
plt.title('Linear Interpolation at x = 0.5 and x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Section 17.3

Question: Use the previous data to plot the cublic spline interpolated points for the same x1 and x2 values.

```python
from scipy.interpolate import CubicSpline
import numpy as np

x,y = [0,1,2],[5,2,4]

function = CubicSpline(x,y, bc_type = 'natural')
xn = np.linspace(0, 2, 100)
yn = function(xn)
y_hat1 = function(0.5)
y_hat2 = function(1.5)

plt.figure(figsize = (10,10))
plt.plot(xn, yn, 'b')
plt.plot(x, y, 'ob')
plt.plot(0.5, y_hat1, 'ro')
plt.plot(1.5, y_hat2, 'ro')
plt.title('Cubic Spline Interpolation at x = 0.5 and x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
s = ("The actual interpolated value for x1 is " + repr(y_hat1) + " and the actual interpolated value for x2 is " + repr(y_hat2) + "." )
print(s)
```
