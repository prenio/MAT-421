<u>Question:</u> The differential equation df(t)dt=e<sup>t</sup>
 with initial condition f<sub>0</sub>=1
 has the exact solution f(t)=e<sup>t</sup>
. Approximate the solution to this initial value problem between 0 and 1 in increments of 0.1 using the Explicity Euler Formula and plot it.

<p>Import needed libraries</p>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-poster')

# Create our ODE for the problem
function = lambda t, s: np.exp(t)

# Set the rest of the variables
step = 0.1
t = np.arange(0, 1+step, step)
init = 1

# Euler Method
s = np.zeros(len(t))
s[0] = init

for i in range(0, len(t)-1):
  s[i+1] = s[i] + step*function(t[i], s[i])
```

```python
# Plot the result
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.title('Approximate Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('function(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

<u>Question:</u> Plot the difference between the approximated solution and the exact solution from the previous problem.

```python
# Plot the exact solution with the approximated solution
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, np.exp(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('function(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

<u>Question:</u> Using the same ODE and initial conditions, find the approximated solution using a step size of 0.01 instead. Plot the difference between the approximated and exact solutions.

```python
# The function is already set, just need to set
# the step size and calculate everything again

step = 0.01
t = np.arange(0, 1+step, step)
init = 1

# Euler Method
s = np.zeros(len(t))
s[0] = init

for i in range(0, len(t)-1):
  s[i+1] = s[i] + step*function(t[i], s[i])
```
