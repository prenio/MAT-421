Question: Compute and plot the Lagrange polynomial for the data set x = [0,1,2], y = [5,2,4] and verify that it goes through each of the data points.

Import the necessary libararies.

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

p1(x) = (x-x2)(x-x3)/(x1-x2)(x1-x3)
      = (x-1)(x-2)/(0-1)(0-2) = .5(x^2 -3x + 2)
<br><br>
p2(x) = (x-x1)(x-x3)/(x2-x1)(x2-x3)
      = (x-0)(x-2)/(1-0)(1-2) = -1(x^2 - 2x)
<br><br>
p3(x) = (x-x1)(x-x2)/(x3-x1)(x3-x2)
      = (x-0)(x-1)/(2-0)(2-1) = .5(x^2 - x)

We can now plot the polynomials as we have found them.

```python
x,y = [0,1,2], [5,2,4]
p1_cof = [1,-1.5,.5]
p2_cof = [0,2,-1]
p3_cof = [0,-.5,.5]

p1 = poly.Polynomial(p1_cof)
p2 = poly.Polynomial(p2_cof)
p3 = poly.Polynomial(p3_cof)
```

```python
lagrange = (y[0]*p1) + (y[1]*p2) + (y[2]*p3)

x_new = np.arange(-2.0, 4.0, 0.1)

fig = plt.figure(figsize = (10,10))
plt.plot(x_new, lagrange(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We can use the lagrange method in scipy to verify this.

```python
from scipy.interpolate import lagrange

function = lagrange(x, y)

fig = plt.figure(figsize = (10,10))
plt.plot(x_new, function(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
