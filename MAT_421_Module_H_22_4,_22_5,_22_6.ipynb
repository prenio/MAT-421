<u>Question:</u> Consider the ODE<br>
        dS(t)/dt = sin(t)<br>
for an initial value S<sub>0</sub> = 0. The exact solution to this problem is S(t) = -cos(t). Use the solve_ivp method to approximate the solution to the initial value problem over the interval |0,Ï€|.

<p>Import needed libraries</p>

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

plt.style.use('seaborn-poster')

function = lambda t, s: np.sin(t)

t_eval = np.arange(0, np.pi, 0.1)
solution = solve_ivp(function, [0, np.pi], [0], t_eval=t_eval)

print(solution)
```

<u>Question:</u> Plot the difference between the approximated solution versus the exact solution and the relative error over time from the previous problem.

```python
plt.figure(figsize = (12, 4))
plt.subplot(121)
plt.plot(solution.t, solution.y[0])
plt.xlabel('t')
plt.ylabel('S(t)')
plt.subplot(122)
plt.plot(solution.t, solution.y[0] + np.cos(solution.t))
plt.xlabel('t')
plt.ylabel('S(t) + cos(t)')
plt.tight_layout()
plt.show()
```
